```python
data = []
for i in range(5):
    print(i)
    index_str = str(i+1)
    ar = df[index_str].values.tolist()
    buf = []
    for x in ar:
        x = float(x)
        buf.append(x)
    data.append(buf)
```

```python
def find_hypothesis_testing(matrix):
    row = len(matrix)
    col = 0
    multotal = False
    total = []
    sample_size = []
    for i in range(len(matrix)):
        col = len(matrix[i])
        sample_size.append(sum(matrix[i]))
        if not multotal:
            total = [0]*len(matrix[i])
            multotal = True
        for j in range(len(matrix[i])):
#             print(matrix[i][j])
            total[j] += matrix[i][j]
    
    all_total = sum(total)
#     print(all_total)
    expected = []
    for i in range(row):
        buf = []
        for j in range(col):
#             print(type(sample_size[i]))
            buf.append(sample_size[i]*total[j]/all_total)
#             print(sample_size[i]*total[j]/all_total)
        expected.append(buf)
#     print(expected)
    
    test_statistic = 0
    for i in range(row):
        for j in range(col):
            n = matrix[i][j]
            e = expected[i][j]
            test_statistic += ((n-e)**2)/e
    return test_statistic
```

```python
n = 5
hypothesis_testing = {}
for i in range(n):
    for j in range(i+1,n):
        for k in range(j+1,n):
            matrix = []
            matrix.append(data[i])
            matrix.append(data[j])
            matrix.append(data[k])
            key_str = 'book ('+ str(i+1) + ',' + str(j+1) + ',' + str(k+1)+')'
            hypothesis_testing[key_str] = find_hypothesis_testing(matrix)
```
