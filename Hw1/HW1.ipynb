```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hw1_data.csv')
df
```

```python
df['%profit']=df.revenue/df.budget
df['%profit'].fillna(0, inplace=True)
df
```

```python
df[pd.isnull(df).any(axis=1)]
```

```python
data = pd.DataFrame(df.groupby('genres')['vote_average','budget','%profit'].apply(np.mean))
data = data.sort_values(by='vote_average',ascending=False)
data
```
