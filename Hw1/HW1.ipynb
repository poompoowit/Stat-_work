```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hw1_data.csv')
df
```

```python
df['%profit']=df.revenue/df.budget
df['%profit'].fillna(0, inplace=True)
df
```

```python
df = df[df.budget!=0]
df = df[df.revenue!=0]
df
```

```python
data = pd.DataFrame(df.groupby('genres')['vote_average','budget','%profit'].apply(np.mean))
data = data.sort_values(by='vote_average',ascending=False)
data = data.astype(float)
data
```

```python
f = plt.figure()
plt.ticklabel_format(style = 'plain')
data.plot(kind='bar',y='budget', ax=f.gca())
```

```python
f = plt.figure()
plt.ticklabel_format(style = 'plain')
data.plot(kind='bar',y='%profit', ax=f.gca())
```

```python
f = plt.figure()
plt.ticklabel_format(style = 'plain')
data.plot(kind='bar',y='vote_average', ax=f.gca())
```
